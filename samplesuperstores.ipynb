```python
# Visualize data as an interactive treemap
# Reference: https://plotly.com/python-api-reference/generated/plotly.express.treemap.html
fig = px.treemap(
  data_frame = df,                        # data frame
  path = ['Category','Sub-Category' ],  # list of column names representing the hierarchy 
  values = 'Sales',                      # column name containing values 
  height = 700,                           # height of the treemap,
  title = 'Visualizing the Fortune 500'   # title of the plot
)
fig.update_layout(margin = dict(l=0, r=0, b=0))
fig.show(config = {"displayModeBar": True})
```

```python
Z=list(set(df['Region']))
W=[]
profit=[]
quantity=[]
for i in Z:
    x=df.loc[df['Region']==i].Profit.sum()
    y=df.loc[df['Region']==i].Quantity.sum()
    profit+=[(x)/800]
    quantity+=[(y)]
    W+=[(x/y)]
fig = go.Figure(data=[go.Scatter(
    x=Z,y=W,
    mode='markers',
    marker=dict(
        color=quantity,
        size=profit,
        showscale=True
        )
)])
fig.show()
```

```python
E=list(set(df['Sub-Category']))
F=[]
profit=[]
quantity=[]
for i in E:
    x=df.loc[df['Sub-Category']==i].Profit.sum()
    y=df.loc[df['Sub-Category']==i].Quantity.sum()
    profit+=[(x+20000)/1000]
    quantity+=[(y)]
    F+=[(x/y)]
print(profit)
fig = go.Figure(data=[go.Scatter(
    x=E,y=F,
    mode='markers',
    marker=dict(
        color=quantity,
        size=profit,
        showscale=True
        )
)])
fig.show()
```
